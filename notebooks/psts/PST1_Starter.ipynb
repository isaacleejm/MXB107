# Dataset
You can load the Star Trek episode dataset that will be analysed in Problem Solving Task 1 with the following R code.

````r
data = read.csv("./datasets/student_wellness_dataset.csv")

head(data)
````

You are now ready to get started working on the Problem Solving Task!

Here is some more information about the dataset.  A sample of students is taking from a particular Univeristy to investigate the impact of student wellness on exam performance.  Each row in the dataset corresponds to a student.  The dataset has the following measurements:


*   Faculty: indicates which Faculty a student belongs to
*   YearLevel: indicates what year level the student is at
*   ExerciseGroup: indicates how much a student exercises as a categorical variable
*   StudyHours: indicates how long a student studies each week on average
*   SleepHours: indicates how long a student sleeps each night on average
*   StressScore: indicates a stress score for a student
*   ExamScore: indicates the score a student got on the exam


````r
hist(data$StudyHours,
     main="Histogram of Study Hours",
     xlab="Study Hours")
````

````r
firstYear_StressScore = subset(data$StressScore, data$YearLevel == "1st Year")
secondYear_StressScore = subset(data$StressScore, data$YearLevel == "2nd Year")
thirdYear_StressScore = subset(data$StressScore, data$YearLevel == "3rd Year")
forthYear_StressScore = subset(data$StressScore, data$YearLevel == "4th Year")

boxplot(firstYear_StressScore, secondYear_StressScore, thirdYear_StressScore, forthYear_StressScore,
        main = "Stress Scores by Year Level",
        names = c("1st Year", "2nd Year", "3rd Year", "4th Year"),
        xlab = "Year Level",
        ylab = "Stress Score"
        )
````

````r
plot(data$StudyHours, data$ExamScore,
     main = "Exam Score against the Study Hours",
     xlab = "Study Hours",
     ylab = "Exam Scores"
     )
````

````r
data$HighExerciseGroup = ifelse(data$ExerciseGroup == "High", 1, 0)
Pr_HighExerciseGroup = sum(data$HighExerciseGroup) / length(data$HighExerciseGroup)
print(paste("1. Probability that randomly selected student is in the high exercise group =", Pr_HighExerciseGroup))


data_FirstYear = subset(data, data$YearLevel == "1st Year")
Pr_HighExerciseGroup_FirstYear = sum(data_FirstYear$HighExerciseGroup) / length(data_FirstYear$HighExerciseGroup)
print(paste("2. Probability that a randomly selected first year student is in the high exercise group =", Pr_HighExerciseGroup_FirstYear))

data$StudiesMoreThan30Hours = ifelse(data$StudyHours > 30, 1, 0)
Pr_StudiesMoreThan30Hours = sum(data$StudiesMoreThan30Hours) / length(data$StudiesMoreThan30Hours)
print(paste("3. Probability that a randomly selected student studies more than 30 hours in a week = ", Pr_StudiesMoreThan30Hours))
````
