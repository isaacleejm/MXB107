## Question 1
Weather bureau records show that for a certain type of weather pattern, the chance of rain at dam A is 40\% and at dam B is 75\%, and the chance of rain at least one of the dams is 80\%. If this weather pattern brings rain to dam B, what is the chance it will also bring rain to dam A?


$ Pr(A) = 0.4 $

$ Pr(B) = 0.75 $

$ Pr(A∪B) = 0.8 $

$  Pr(A|B) = \frac{Pr(AB)}{Pr(B)} $

$  Pr(A|B) = \frac{Pr(A) + Pr(B) - Pr(A∪B)}{Pr(B)} $

$  Pr(A|B) = \frac{0.4 + 0.7 - 0.8}{0.75} $

$  Pr(A|B) = 0.467 $





<details>
<summary>▶️ Click to show the solution</summary>

Let $R_A$ be the event that it rains at dam A and $R_B$ be the event that it rains at dam B. Then
			$$ Pr(R_A) = 0.4,\, Pr(R_B)=0.75,\,Pr(R_A\cup R_B)=0.8 $$

We want
			$$\begin{align*}
				Pr(R_A|R_B) &= \frac{Pr(R_AR_B)}{Pr(R_B)} \quad  \text{ conditional rule} \\
				&= \frac{Pr(R_A)+Pr(R_B)-Pr(R_A\cup R_B)}{Pr(R_B)} \quad \text{ addition rule} \\
				&=\frac{0.4+0.75-0.8}{0.75}\\
				&\approx  0.467.
			\end{align*} $$

</details>

b)  If the test says that the bacteria are not present in a sample, what is the probability that the bacteria are present?

c) The testing is extended to include sources of water that feed into dams. In such sources it is known that the bacterial strain occurs in 3\% of the water. Twice as much testing is carried out for dams as for the other sources. In what proportions of tests that return a negative result (that is, that say the bacteria are not present) are the bacteria actually present?

<details>
<summary>▶️ Click to show the solution</summary>

Let $B$ be the event that bacteria is present, and let $T$ be the event that the test says bacteria is present.

Probabilities:
			$$P(B) = 0.02,\, P(T|B)=0.95,\, P(T|\overline{B})=0.1.$$

a) \begin{align*}
				P(\overline{B}|T) &= \frac{P(T|\overline{B})P(\overline{B})}{P(T)}&&\text{ Bayes' rule} \\
					&=\frac{P(T|\overline{B})P(\overline{B})}{P(T|\overline{B})P(\overline{B})+P(T|B)P(B)}&& \text{law of total probability} \\
					&= \frac{0.1\times0.98}{0.1\times0.98+0.95\times0.02}\\
					&\approx 0.8376.
				\end{align*}

b)  Note from the complement rule that $P(\overline{T})=1-P(T) = 1-0.117 = 0.883$
			\begin{align*}
				P(B|\overline{T}) &= \frac{P(\overline{T}|B)P(B)}{P(\overline{T})}&&\text{ Bayes' rule}\\
					&= \frac{0.05\times0.02}{0.883}\\
					&\approx 0.00113
			\end{align*}

c)  Let D be the event that water comes from the dam. Then

$$P(D)=2/3, \text{ and } P(\overline{D})=1/3.$$

To calculate $P(B)$ use the law to total probability.

$$P(B) = P(B|D)P(D)+P(B|\overline{D})P(\overline{D}) = 0.02\times2/3+0.03\times1/3\approx0.0233$$

Now we can determine the probability that the bacteria are present when the test says they are not present.

\begin{align*}
					P(B|\overline{T}) &= \frac{P(\overline{T}|B)P(B)}{P(\overline{T})}&&\text{ Bayes' rule}\\
						&= \frac{P(\overline{T}|B)P(B)}{P(\overline{T}|B)P(B)+P(\overline{T}|\overline{B})P(\overline{B})}&&\text{ law of total probability} \\
						&\approx \frac{0.05\times0.0233}{0.05\times0.0233+0.9\times(1-0.0233)}\\
						&\approx 0.00132.
				\end{align*}

Based on newspaper circulation figures and previous survey information, the marketing company states that the probability that an arbitrary person sees the ads is 60\%. During the newspaper ads campaign, the buyers of the product were asked if they'd seen the ads, 75\% replied that they had. Assume that for those who did not see the ads, the probability of buying the product is still 5\%.

a) Show that the probability that a person who saw the ads buys the product is exactly 10%.

b) Why is $Pr(B)\neq 0.05$, where $B$ is the event "customer buys product"?

<details>
<summary>▶️ Click to show the solution</summary>
Let $S$ be the event of a customer seeing the product and let $B$ be the event that a customer buys the product.

Given Probabilities:
			\begin{align*}
				P(S) &= 0.6 &
				P(S|B) &= 0.75\\
				P(B|\overline{S}) &= 0.05 &
				P(B|S) &= 0.1
			\end{align*}

a)
				\begin{align*}
					P(B|S) &= \frac{P(S|B)P(B)}{P(S)} && \text{ Bayes' rule}\\
					P(B) &= P(BS)+P(B\overline{S}) &&\text{ law of total probability} \\
						&= P(B|S)P(S)+P(B|\overline{S})P(\overline{S})\\
						&= 0.6P(B|S)+0.02\\
					P(B|S) &=\frac{0.75(0.6P(B|S)+0.02)}{0.6}\\
					P(B|S)-0.75P(B|S) &= \frac{0.75\times0.02}{0.6}\\
					P(B|S) &= \frac{0.025}{0.25}\\
						&= 0.1
				\end{align*}


b) The reason why $Pr(B)\neq 0.05$ during or after the campaign is because prior to the campaign the entire population had "not seen the ad". During and after the campaign, the population is divided into those who have not seen the ad and those who have seen the ad and those who have seen the ad, will have an increased probability of buying the product. We should expect $Pr(B)>0.05$ after the campaign, and that is exactly what we see with the law of total probability ($Pr(B)=0.08$).

### **Question 4**

We have seen the classical definition of probability based on counting. There is another important (classical) definition — the frequentist definition — based on relative frequency. It defines the probability of an event as the limiting value of the proportion of times the event occurs as the number of trials becomes very large:

$$
\Pr(A) = \lim_{n \to \infty} \frac{N_A(n)}{n}
$$

where $N_A(n)$ is the number of times event $A$ occurs in $n$ trials.


We know that the probability of getting `Heads` (1) by tossing a fair coin is 0.5. Similarly, the probability of getting `Tails` (0) is 1 - 0.5 = 0.5. To simulate this coin tossing experiment `n` times, we could run `sample(0:1, size = n, replace = T)`.

````r
n = 10
x = sample(0:1, size = n, replace = T) #Head = 1; Tail = 0
print(x)
mean(x) #The relative frequency of Head
````

Assume that you want to toss a coin `n` times then calculate the relative frequency of `Heads` and repeat the process `nRepeats` times. To perform this experiment in R, we can do the following steps:

- Create an empty numeric vector of length `nRepeats` to store the frequency of `Heads`.

````r
nRepeats = 100
headFreqn10 = numeric(nRepeats) #Here, n = 10.
````

- We can use a `for` loop to repeat the process `nRepeats` times and save the results to `headFreqn10`.

````r
for(i in 1:nRepeats){ #Repeat for i varying from 1 to nRepeats
   x = sample(0:1, size = 10, replace = T) #Toss a coin 10 times
   headFreqn10[i] = mean(x) #Compute the relative frequency of `Head` in `x` and save the value to the i-th value in `headFreqn10`
}
````

#### **Question 4.1**

Follow the steps above to create new vectors named `headFreqn100`, `headFreqn1000`, and `headFreqn10000` that contain the relative frequencies of heads in 100, 1000, and 10000 tosses, respectively (repeated 100 times).
